```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from natsort import natsorted 
from scipy import stats

%matplotlib inline
```

```python
datadir = '/home/mei/nas/docker/dataset/EICU/eicu-collaborative-research-database-2.0/'
porcesseddir = '/home/mei/nas/docker/processedData/'
```

```python
df = pd.read_csv(datadir + 'respiratoryCharting.csv.gz', compression='gzip',dtype = {'respchartvalue': 'str'})
```

```python
def round_up(x, base=5):
    return base * round(x/base)
```

```python
df_patient = pd.read_csv(porcesseddir + 'selected_patients.csv')
df_selected_res = pd.merge(df_patient, df, on='patientunitstayid')
```

```python
df = df_selected_res.copy()
df = df.set_index(['patientunitstayid', 'respchartoffset'])
df = df.drop(df.index[np.where(df.index.get_level_values('respchartoffset') < 0)])
df = df.drop(columns = ['respchartid', 'respchartentryoffset'])
df.sort_index(level = ['patientunitstayid', 'respchartoffset'], inplace = True)

df.head()
```

```python
selected_res_p=len(list(df.index.get_level_values('patientunitstayid').unique()))
selected_res_x=len(df)
print("There are {} unique patientunitstayid in the selected patients with the total {} records.".format(selected_res_p,selected_res_x))
```

```python
df_care = pd.read_csv(datadir + 'respiratoryCare.csv.gz', compression='gzip',low_memory=False)
df_selected_res_care = pd.merge(df_patient, df_care, on='patientunitstayid')
```

```python
df_c=df_selected_res_care.copy()
df_c = df_c.set_index(['patientunitstayid', 'respcarestatusoffset'])
df_c= df_c.drop(df_c.index[np.where(df_c.index.get_level_values('respcarestatusoffset') < 0)])
df_c = df_c.drop(columns = ['respcareid'])
df_c.sort_index(level = ['patientunitstayid', 'respcarestatusoffset'], inplace = True)

df_c.head()
```

```python
selected_res_c_p=len(list(df_c.index.get_level_values('patientunitstayid').unique()))
selected_res_c_x=len(df_c)
print("There are {} unique patientunitstayid in the selected patients with the total {} records.".format(selected_res_c_p,selected_res_c_x))
```
